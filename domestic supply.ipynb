```python
#importing data which contains price received by each state for fresh market blueberries in respective marketing year
data=pd.read_csv('/Users/ajitkhanal/Downloads/Research/Blueberry/Data/dpricestate.csv')
```

# Data Exploration and Cleaning

```python
data.head()
```

```python
#deleting columns that are not necessary
data.drop(['Program','Period','Geo Level','Domain Category','Domain'],axis=1,inplace=True)
```

```python
data.dtypes
```

```python
data.head(5)
```

```python
#drop rows containing florida
data.drop(data[((data.State=='FLORIDA'))].index, inplace = True)
```

```python
data.head(2)
```

```python
#we delete the data which has zero as the price received to get more precise value of average price received
data1=data.drop(data[((data.Value==0))].index)
```

```python
data.State.value_counts()
```

```python
#exploring the average price received by states
data_g=data1.groupby('State',as_index=False)['Value'].mean()
```

```python
sns.lineplot(x='Year',y='Value',hue='State',data=data)
```

```python
data_g.head()
```

```python
data_g.rename({'Value': 'Price Received($/lb)'}, axis=1, inplace=True)
```

```python
data_g
```

```python
 sns.barplot(x="State", y="Price Received($/lb)", data=data_g)
```

#### Now we replace the rows with zeros with average state received price 

```python
#mean for price recieved by california is 3.049167
mean_value=3.049167
data['Value']=data.Value.replace(0,mean_value)
```

```python
#checking if it worked
data.iloc[30:40,0:4]
```

```python
#now we groupby year to get the average price received 
data_avpr=data.groupby('Year',as_index=False)['Value'].mean()
```

```python
data_avpr.tail()
```

```python
data_avpr.dtypes
```

```python
data_avpr['Years']=pd.to_datetime(data_avpr['Year'],format='%Y')
```

```python
data_avpr.dtypes
```

```python
sns.lineplot(x='Years',y='Value',data=data_avpr)
```

```python
data_avpr.head()
```

```python
#now we remove the years 2002 through 2006 to match the domestic supply data we have
#this is because we have production data on states from 2007
data_avpr1=data_avpr.drop(data_avpr.index[0:5])
```

```python
data_avpr1
```

```python
#now we merge this data frame with domestic supply data
supply_d = d_supply.merge(data_avpr1, on='Year', how='left')
```

```python
supply_d.head()
```

```python
#droping the extra Years column
supply_d.drop(['Years'],axis=1,inplace=True)
```

```python
#Renaming 
supply_d.rename({'Value':'Average Domestic Price($/lb)'},axis=1,inplace=True)
```

```python
#changing units to million pounds
supply_d['Production(Million Pounds)']=supply_d['Production']/1000000
```

```python
supply_d['Exports(Million Pounds)']=supply_d['Exports(lb)']/1000000
```

```python
supply_d.head(2)
```

```python
supply_d['Price']=supply_d['Average Domestic Price($/lb)']*supply_d['Average Domestic Price($/lb)']
```

```python
supply_d.head()
```

```python
supply_d.to_csv('domesticsupply.csv')
```
