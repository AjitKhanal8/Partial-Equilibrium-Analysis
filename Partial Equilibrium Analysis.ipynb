```python
#simple simulation
steps=3000
sig=5
Ei=3
Ed=1
thi=-1.0
simlst=[]
z=0
g=simd['g'].iat[z]
for st in range(10):
    year=2012+st
    invalue=simd['A_iamt'].iat[st]
    A_iamt=simd['A_iamt'].iat[st]
    m__d=simd['M_d'].iat[st]
    m__i=simd['M_i'].iat[st]
    g=simd['g'].iat[st]
   
    for zt in range(1,steps+1):
        A_iamt=A_iamt*(1+g)
        pchange=(((A_iamt-invalue)/invalue)*(((m__d*sig-m__i*sig)/Ei)/(Ed-m__d*sig+m__d*thi+sig)))
        simlst.append([year,pchange])
        
        
        
```
