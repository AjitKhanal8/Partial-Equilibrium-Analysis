```python
imports=df2.dropna()
```

```python
imports
```

```python
imports['Time'] = imports['Time'].dt.year
```

```python
imports
```

```python
imports['Imports']=imports['Quantity1']*2.205
```

```python
imports
```

```python
imports['Million Pounds']=imports['Imports']/1000000
```

```python
imports
```

```python
import seaborn as sns
```

```python
imports['Year']=pd.to_datetime(imports['Time'],format='%Y')
```

```python
#to save the figure we need pylpt
import matplotlib.pyplot as plt
```

```python


sns.set_theme(style='darkgrid')
fig=sns.lineplot(x='Year',y="Million Pounds",hue="Country",palette= ['r','g','b','c','y'],data=imports)
plt.savefig('output.png')
```

```python
#we want greater figure size
from matplotlib import rcParams
```

```python
rcParams['figure.figsize']=8,5
sns.set_theme(style='darkgrid')
fig=sns.lineplot(x='Year',y="Million Pounds",hue="Country",palette= ['r','g','b','c','y'],data=imports)
plt.savefig('output2.png')
```
